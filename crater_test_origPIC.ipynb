```python
k = 15
embed = embedding9
neighs = 300

Kmean = KMeans(n_clusters=k)
label = Kmean.fit_predict(embed)
plt.scatter(embed[ : , 0], embed[ : , 1], s =.1)
# print("teeeh")
print(label.size)
# color = iter(plt.rainbow(np.linspace(0, 1, k)))
# 'hot' used for centroids
cmap = plt.get_cmap('Greens', k)
cmaper = plt.get_cmap('gray', k+1)
u_labels = np.unique(label) #is used to assign each point to its color

for i in range(k):
    plt.scatter(embed[label == i][ : , 0], embed[label == i][ : , 1], s =.1, color=cmap(i))
    # plt.scatter(embed[label == i][ : , 0], embed[label == i][ : , 1], s =.1, c = embed.label, cmap = 'prism')
    plt.scatter(Kmean.cluster_centers_[i][0], Kmean.cluster_centers_[i][1], s=50, color=cmaper(i), marker='s')
plt.gca().set_aspect('equal', 'datalim')
plt.title('n_neighbors= '+ str(neighs)+', min_dist=0.001,metric=correlation')
plt.show()
```
